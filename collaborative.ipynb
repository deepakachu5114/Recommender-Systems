- This code uses the TransactionEncoder from the mlxtend library to perform one-hot encoding on a transactional dataset (transactions). The resulting binary matrix is converted into a pandas DataFrame (encoded_df), where each row represents a transaction, and each column corresponds to an item, with values indicating the presence or absence of the item in the transaction.

```python
from mlxtend.preprocessing import TransactionEncoder
encoder = TransactionEncoder()

encodings = encoder.fit(transactions).transform(transactions)

encoded_df = pd.DataFrame(encodings, columns=encoder.columns_)
encoded_df
```

- This code defines a function market_basket_analysis that performs market basket analysis using the Apriori algorithm from the mlxtend library. The function takes a transactional dataset (data), minimum support (min_support), and minimum confidence (min_confidence) as parameters and returns the generated association rules.


```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

def market_basket_analysis(data, min_support, min_confidence):
    # Apply Apriori algorithm
    frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)
    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

    return rules
```

- The generated rules are printed below, as one can see, these rules do make sense

```python
rules = market_basket_analysis(encoded_df, 0.005, 0.1)
rules
```

```python
# Assuming your DataFrame is named 'df'
# Grouping the DataFrame by 'Customer ID' and aggregating the 'Product Names', 'Quantities', and 'Unit Prices'
user_orders = df_b7_user.groupby('Customer ID').agg({
    'Customer ID': lambda x: x.iloc[0],  # Select the first value of 'Customer ID' for each group
    'Product Names': lambda x: list(x),
    'Quantities': lambda x: list(x),
    'Unit Prices': lambda x: list(x)
}).reset_index(drop=True)


# Displaying the grouped data
print(user_orders)
```

```python
user_level_transactions = user_orders['Product Names']
user_level_transactions
```

```python
series = user_level_transactions

# Extract unique dishes from the series
unique_dishes = list(set([dish for dishes_list in series for dish in dishes_list]))

# Create a dictionary to map each dish to its index in the matrix
dish_index_map = {dish: i for i, dish in enumerate(unique_dishes)}

# Initialize the matrix with zeros
matrix = np.zeros((len(series), len(unique_dishes)))

# Iterate over each element (list of dishes) in the series
for i, dishes_list in enumerate(series):
    # Iterate over each dish in the list of dishes for that user
    for dish in dishes_list:
        # Get the index of the dish in the matrix
        dish_index = dish_index_map[dish]
        # Increment the count for that user and that dish
        matrix[i, dish_index] += 1

# Convert the matrix to a DataFrame with appropriate column names
matrix_df = pd.DataFrame(matrix, columns=unique_dishes)

# Print or return the resulting DataFrame
matrix_df
```

```python
user_ids = user_orders.iloc[:, 0]
user_interaction_matrix = pd.concat([user_ids, matrix_df], axis=1)
user_interaction_matrix
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def find_top_similar_users(user_id, interaction_df, n=5):
    # Get the index of the target user
    user_index = interaction_df[interaction_df.columns[0]] == user_id

    # Extract the interaction data excluding the user ID column
    interaction_data = interaction_df.iloc[:, 1:]

    # Calculate cosine similarity between the target user and all other users
    similarities = cosine_similarity(interaction_data.values[user_index], interaction_data.values)[0]

    # Get the indices of top N similar users (excluding the target user)
    top_similar_indices = np.argsort(similarities)[-n-1:-1][::-1]

    # Get the user IDs of top N similar users
    top_similar_users = interaction_df.iloc[top_similar_indices, 0]

    return top_similar_users


user_id = 20104
top_similar_users = find_top_similar_users(user_id, user_interaction_matrix, n=5)
print("index\tuserid")
print(top_similar_users)

```

```python
import pandas as pd

def get_top_k_recommendations(user_interaction_matrix, top_similar_users, k=5):
    top_k_recommendations = {}

    for user_id in user_interaction_matrix['Customer ID']:
        # Get the preferences of the given user
        user_preferences = user_interaction_matrix.loc[user_interaction_matrix['Customer ID'] == user_id]
        
        # Filter out items that the given user has already interacted with
        user_preferences = user_preferences[user_preferences.columns[1:]].T
        user_preferences = user_preferences[user_preferences[user_preferences.columns[0]] == 0]
        
        # Initialize a dictionary to store aggregated preferences of similar users
        similar_users_preferences = {}
        
        # Iterate over top similar users and aggregate their preferences
        for similar_user_id in top_similar_users:
            similar_user_preferences = user_interaction_matrix.loc[user_interaction_matrix['Customer ID'] == similar_user_id]
            similar_user_preferences = similar_user_preferences[similar_user_preferences.columns[1:]].T
            similar_user_preferences.columns = ['Preference']
            
            # Filter out items that the similar user has interacted with
            similar_user_preferences = similar_user_preferences[similar_user_preferences['Preference'] > 0]
            
            # Update similar users preferences dictionary
            for index, row in similar_user_preferences.iterrows():
                if index not in similar_users_preferences:
                    similar_users_preferences[index] = 0
                similar_users_preferences[index] += row['Preference']
        
        # Convert the dictionary to a DataFrame
        similar_users_preferences_df = pd.DataFrame(list(similar_users_preferences.items()), columns=['Item', 'Aggregate_Preference'])
        
        # Sort the DataFrame by aggregate preference in descending order
        similar_users_preferences_df = similar_users_preferences_df.sort_values(by='Aggregate_Preference', ascending=False)
        
        # Filter out items that the given user has not interacted with
        recommendations = similar_users_preferences_df[~similar_users_preferences_df['Item'].isin(user_preferences.index)]
        
        # Get top k recommendations
        top_k_recommendations[user_id] = recommendations.head(k)['Item'].tolist()

    # Convert the dictionary to a DataFrame
    top_k_recommendations_df = pd.DataFrame.from_dict(top_k_recommendations, orient='index').reset_index()
    top_k_recommendations_df.columns = ['Customer ID'] + [f'Top_{i}_Recommendation' for i in range(1, k+1)]
    
    return top_k_recommendations_df


top_k_recommendations_df = get_top_k_recommendations(user_interaction_matrix, top_similar_users, k=5)
top_k_recommendations_df

```
